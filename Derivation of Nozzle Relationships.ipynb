# Derrivation of Nozzle Relationships


## Enthalpy and Ideal Gas Law


By Definition, enthalpy is defnied as internal energy + pressure-volume product  

> h = u + pv
    
    
The ideal gas law states:

> pv = Rt
    
where R = Ro / MW
   
Substituting, 

> h = u + Rt
    
differentiating

> dh = du + d(Rt)
    
>        = du + RdT + TdR    since R is a constant, TdR -> 0 
        
> dh = du + RdT    
    
For an ideal gas with constant specific heats, [e.g. h = c<sub>p</sub>(T<sub>2</sub> - T<sub>1</sub>)]    

c<sub>p</sub> dT = c<sub>v</sub>dT + RdT  

        
Divide by dT  
  
> c<sub>p</sub> = c<sub>v</sub> + R    
        
or  
        
> c<sub>p</sub> - c<sub>v</sub> = R   
    
    
Define specific heat ratio:  
 
> k = c<sub>p</sub> / c<sub>v</sub>  

> c<sub>p</sub> * (1 - 1/k) = R

> c<sub>p</sub> = k * R / (k-1)


## First Law of Thermodynamics

> Q' - W' = m' [H + KE + PE]<sub>2</sub>  - m' [H + KE + PE]<sub>1</sub>

### Simplifying conditions:

- Adiabatic, Q' = 0

- Rigid and no shaft work, W' = 0

- No change in elevation, PE = 0

- Velocity in the nozzle chamber be small, KE<sub>1</sub> = 0

- Steady mass flow (no accumulation), m'<sub>1</sub> = <sub>2</sub>         

> 1/2 V<sub>J</sub><sup>2</sup> = h<sub>i</sub> - h<sub>e</sub>
> V<sub>J</sub><sup>2</sup> = 2 * ( h<sub>i</sub> - h<sub>e</sub> )   

> V<sub>J</sub><sup>2</sup> = c<sub>p</sub> * [T<sub>c</sub> - T<sub>e</sub>]

### Nozzle exit velocity by substituting results of ideal gas law:

> V<sub>J</sub><sup>2</sup>  =  $\frac{2kRoTc}{MW(k-1)}$ [Tc - Te]

> V<sub>J</sub><sup>2</sup>  =  $\frac{2kRoTc}{MW(k-1)}$ [1 - (Te/Tc)]  

> V<sub>J</sub>  =  $\sqrt{\frac{2kRoTc}{MW(k-1)}$ [1 - (Te/Tc)]}$

### Pressure and temperature relationships:  

Proceeding from Tds = dh - dp / $\rho$  

dividing by T

> ds = dh / T - dp / $\rho$ T

for isentropic flow, ds = 0, thus

> dh / T = dp / $\rho$ T
  
for a perfect gas, p / R = $\rho$T  

substituting, 

> dh / T = dp / (p / R) = R dp / p

Further substituting dh = c<sub>p</sub>dT

> c<sub>p</sub>dT / T = Rdp / p

Integrating: 


> $\int\limits_1^2{\frac{Cp dT}{T}}t$ = R$\int\limits_1^2{\frac{ dp}{p}}$


> ln$\frac{T2}{T1}$ = $\frac{R}{Cp}$ln$\frac{P2}{P1}$

> $\frac{R}{Cp}$ = (k-1) / k

> e<sup>(ln$\frac{T2}{T1}$)</sup> = e<sup>(ln$\frac{P2}{P1}$)<sup>(k - 1) / k)</sup></sup>  


> $\frac{T2}{T1}$ = $\frac{P2}{P1}$<sup>(k - 1) / k</sup>

A similar exercised can be conducted for density, $\rho$ to yield:

> $\frac{P2}{P1}$ = ($\frac{\rho2}{\rho1}$)<sup>k</sup>


combining, 

> $\frac{T2}{T1}$ = ($\frac{P2}{P1}$)<sup>(k - 1) / k</sup> = ($\frac{\rho2}{\rho1}$)<sup>(k-1)</sup>
